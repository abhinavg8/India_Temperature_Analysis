```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.style as style

df = pd.read_csv('Min_Max_Seasonal_IMD_2017.csv')

## For Styling..
style.use('bmh')
plt.rcParams['font.family'] = "serif"
```

```python
df.head()
```

```python
years = list(np.linspace(1901,2017,5))
years = [str(int(x)) for x in years]
```

## Seasonal Change in the temperature over the years

```python
fig, axis = plt.subplots(nrows = 4,ncols = 2, figsize = (15,18),gridspec_kw={'hspace': 0.35, 'wspace': 0.5})

for i in range(0,4):
    for j in range(0,2):
        plt.sca(axis[i,j])
        plt.xticks(np.linspace(0,120,5), years)

axis[0,0].plot(df['JAN-FEB - MAX'], c ='red', alpha= 0.50)
axis[0,0].set_title('JAN _FEB - MAX')

axis[0,1].plot(df['JAN-FEB - MIN'],c='blue',alpha= 0.50)
axis[0,1].set_title('JAN _FEB - MIN')

axis[1,0].plot(df['MAR-MAY - MAX'], c = 'red', alpha= 0.50)
axis[1,0].set_title('MAR_MAY - MAX')

axis[1,1].plot(df['MAR-MAY - MIN'],c='blue', alpha= 0.50)
axis[1,1].set_title('MAR_MAY - MIN')

axis[2,0].plot(df['JUN-SEP - MAX'], c ='red', alpha= 0.50)
axis[2,0].set_title('JUN-SEP - MAX')

axis[2,1].plot(df['JUN-SEP - MIN'],c='blue', alpha= 0.50)
axis[2,1].set_title('JUN-SEP - MIN')

axis[3,0].plot(df['OCT-DEC - MAX'], c = 'red', alpha= 0.50)
axis[3,0].set_title('OCT-DEC - MAX')

axis[3,1].plot(df['OCT-DEC - MIN'],c='blue',alpha= 0.50)
axis[3,1].set_title('OCT-DEC - MIN')

for ax in axis.flat:
    ax.set(xlabel='Year', ylabel='Temperature(in Celcius)')
    
fig.suptitle('Seasonal Change in Temperature(1901-2017)', x=0.5, y =0.94 ,size=25)
```

## Annual Change in the temperature over the years

```python
plt.figure(figsize=(10,5))
plt.plot(df['ANNUAL - MAX'],c = 'red',alpha=0.5)
plt.xticks(np.linspace(0,120,5), years)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.title('Change in Maixmum Temperature(1901-2017)')
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['ANNUAL - MIN'],c = 'blue',alpha=0.5)
plt.xticks(np.linspace(0,120,5), years)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.title('Change in Minimum Temperature(1901-2017)')
```

## Mean Change of Temperature over the years

```python
df['Mean'] = (df['ANNUAL - MAX'] + df['ANNUAL - MIN'])/2
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Mean'],c = 'orange')
plt.xticks(np.linspace(0,120,5), years)
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.title('Change in Mean Temperature(1901-2017)')
```
